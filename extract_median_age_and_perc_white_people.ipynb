### This file extracts the median age per state as well as the percentage of white people per state. The data comes from the american community survey.

### Two files are outputed (median_age_per_state.csv, percentage_white_people_per_state.csv)

```python
import pandas as pd
```

```python
df = pd.read_csv('ACSDP1Y2021.DP05-Data.csv')
```

```python
df.head(5)
```

```python
#get descriptions
[i for i in df.loc[0]]
```

```python
#filter columns finishing with a E (for estimate)
filter_col = [col for col in df if col.endswith('E')]
```

```python
len(filter_col)
```

```python
filter_col
```

```python
#create new smaller df with new set of columns
df = df[filter_col]
```

```python
df.head(5)
```

```python
#create list of tuples with column name and first row descritpion
link_first_row_col_names = list(zip(df.columns, df.loc[0]))
```

```python
link_first_row_col_names[:5]
```

```python
median_age_per_state = df[['NAME','DP05_0018E']][1:52]
```

```python
median_age_per_state.head(2)
```

```python
percentage_white_people_per_state = df[['NAME', 'DP05_0037PE']][1:52]
percentage_white_people_per_state.head(2)
```

```python
median_age_per_state = median_age_per_state.rename(columns={'DP05_0018E':'median_age', 'NAME':'State'})
percentage_white_people_per_state=percentage_white_people_per_state.rename(columns ={'DP05_0037PE':'perc_white_people','NAME':'State'})
```

```python
states_abbr = pd.read_csv('states_abbr.csv')
states_abbr.head(2)
```

```python
median_age_per_state=median_age_per_state.merge(states_abbr, how='left', on ='State')
percentage_white_people_per_state=percentage_white_people_per_state.merge(states_abbr, how='left', on='State')
```

```python
median_age_per_state.head(2)
```

```python
median_age_per_state.drop(['Abbrev','State'], axis=1,inplace=True)
percentage_white_people_per_state.drop(['Abbrev', 'State'], axis=1, inplace=True)
```

```python
median_age_per_state.to_csv('median_age_per_state.csv')
```

```python
percentage_white_people_per_state.to_csv('percentage_white_people_per_state.csv')
```
