# This file is used to extract the distance to the closest airport and produces a csv file 

```python
import pandas as pd
import numpy as np

from geopy import distance
```

```python
# Load the training and test data
train = pd.read_csv('train.csv', index_col='Uniq Id')
test = pd.read_csv('test.csv', index_col='Uniq Id')
```

```python
airports = pd.read_csv('airports.csv',index_col= 'IATA')
```

```python
airports
```

```python
test_test = pd.read_csv('test.csv')
```

```python
test_train= pd.read_csv('train.csv')
```

```python
airports_test_test = pd.read_csv('airports.csv')
```

```python
df3 = test_test[['Uniq Id', 'Latitude','Longitude']]
df4 = airports_test_test[['IATA', 'LATITUDE','LONGITUDE']]
df5= test_train[['Uniq Id', 'Latitude','Longitude']]
```

```python
df3 = df3.rename(columns = {'Uniq Id': 'id'})
df4 = df4.rename(columns = {'IATA': 'id'})
df5 = df5.rename(columns = {'Uniq Id': 'id'})
```

```python
df4
```

```python
df3
```

```python
df5
```

```python
dfm_test = pd.merge(df3, df4, how = 'cross', suffixes = ['','_2'])
```

```python
dfm_test['dist'] = dfm_test.apply(lambda r: distance.distance((r['Latitude'],r['Longitude']),(r['LATITUDE'],r['LONGITUDE'])).km * 1000 , axis=1)
```

```python
dfm_train = pd.merge(df5, df4, how = 'cross', suffixes = ['','_2'])
```

```python
dfm_train['dist'] = dfm_train.apply(lambda r: distance.distance((r['Latitude'],r['Longitude']),(r['LATITUDE'],r['LONGITUDE'])).km * 1000 , axis=1)
```

```python
dfm_test #should be 5000*341=1705000
```

```python
dfm_train #should be 20000*341=6'820'000
```

```python
res_test_dist=dfm_test.groupby('id')['dist'].min()
```

```python
res_train_dist=dfm_train.groupby('id')['dist'].min()
```

```python
res_train_dist #should be 20'000
```

```python
res_test_dist.to_csv('Test_set_Airport_dist_column.csv')
```
